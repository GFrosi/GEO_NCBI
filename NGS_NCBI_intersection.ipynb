#FUNCTION to categorize control and trait samples

```python
#function 

def control_trait_class(df):
    

    import re 


    list_categories = []


    patterns = [r"\bcontrol\b",r"\binput\b",r"[-_]?(input)[_-]?", r"[-_]?(Control)[_-]?", 
            r"\bctrl\b", r"[-_]?(ctrl)[_-]?", r"\bctl\b", r"[-_]?(ctl)[_-]?"]

    compiled_combined = re.compile('|'.join(x for x in patterns), re.IGNORECASE)


    for index, row in df.iterrows():
        if re.search(compiled_combined, row['Title']) or re.search(compiled_combined, row['source-cell']):
#         print(row['Title'] ,"::", row['source-cell'])
            list_categories.append('Control')
        

        else:       
#         print(row['Title'])
            list_categories.append('Trait')
    
    return list_categories
       
        
```

#FUNCTION CELL TYPE

```python
def find_cell_type(df, cell_type):
    
    return df[(df['cell type'].str.contains(cell_type,case=False, na=False)) | 
        (df['Title'].str.contains(cell_type,case=False, na=False) | 
        (df['source-cell'].str.contains(cell_type,case=False, na=False)))]
```
