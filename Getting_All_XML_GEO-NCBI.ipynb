```python
import pandas as pd
import numpy as np
import subprocess
import os

pd.set_option('display.max_rows', 20)

df_ftp_address = pd.read_csv('/Users/gfrosi/Documents/Frosi_lab_jacques/XML_TEST/GSE_FTP_Hs_ChipSeq.csv' , sep=',')
df_ftp_address
```

```python
#all XMLs except the 157 missing
def create_dir(path):    
    try:
        os.mkdir(path)
        print("Directory {} done".format(path))
    except OSError:
        print("Directory {} failed".format(path))
            

root_path = '/Users/gfrosi/Documents/Frosi_lab_jacques/XML_TEST/GEO-data-xml'
prefix_dir = 'GEO_'
counter = 1                                                                  

path = os.path.join(root_path , prefix_dir + str(counter))
create_dir(path)


#create the correct link and download the xml files
for index, row in df_ftp_address.iterrows():    
    if (index + 1) % 300 == 0:  
        counter += 1
        path = os.path.join(root_path, prefix_dir + str(counter))
        create_dir(path)    
        
    #sample_name = row["Address"].split("/")
    link = row["Full_Address"] #+ sample_name[6] + "_family.xml.tgz" #we can keep in this form, or get the full address
    subprocess.run(['wget', link, '-P', path])
    
```
