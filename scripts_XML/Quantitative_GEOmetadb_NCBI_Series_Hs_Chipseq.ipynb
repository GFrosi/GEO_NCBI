```python
import urllib.request
import lxml.html as lh
from bs4 import BeautifulSoup
import requests
import re
from time import sleep




regex = re.compile('Samples\s+\(\d+\)')  #get the sample (number)
gse_samples_ncbi =[]

for gse in gse_list_GEOdb:
    url = 'https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=' + gse
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data, 'lxml')
    tr = soup.findAll("tr", {"valign" : "top"}) #ok    

    for td in tr:
        match = re.search(regex, td.text)
        if match:
            gse_samples_ncbi.append([gse, match.group(0)])
    
    sleep(3) #Ok

print(gse_samples_ncbi)
```

```python
# V2
import urllib.request
import lxml.html as lh
from bs4 import BeautifulSoup
import requests
import re
from time import sleep


gse_list_samples_file = open("gse_list_sample_file.tsv", "a+")

regex = re.compile('Samples\s+\(\d+\)')  #get the sample (number)

err_samp = []
index_new_start = 2484
line = ''
for i,gse in enumerate(gse_list_GEOdb[index_new_start:]):
    try:
        url = 'https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=' + gse
        response = requests.get(url)
        data = response.text
        soup = BeautifulSoup(data, 'lxml')
        tr = soup.findAll("tr", {"valign" : "top"}) #ok    
    except:
        print('Error on sample', gse)
        err_samp.append(gse)          # Get problematic samples (try later)
        sleep(60)                     # sleep more so NCBI let me pass
        
    else:
        for td in tr:
            match = re.search(regex, td.text)
            if match:
                line = str(i + index_new_start) +"\t"+ gse +"\t"+ match.group(0) +"\n"
                gse_list_samples_file.write(line)
    
    sleep(4) #Ok

# print(gse_samples_ncbi)
gse_list_samples_file.close()
```
