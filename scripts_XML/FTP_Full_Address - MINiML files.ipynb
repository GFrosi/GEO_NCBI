# Efetch Human and Chip-Seq data


```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 20)
```

```python
#loading the ftp information from efetch table - Chip-Seq and Homo sapiens 
#without pathway
df = pd.read_csv('list_ftp_address.txt', sep=r'\s{2,}', engine='python', header=None, names=['ID'])

df['Address'] = df.ID.str.split("ftp:",expand=True)[1] #impossible split by GEO (There is no patters to GEO string )

#building the first part of address
df['Address'] = 'ftp:' + df['Address'].astype(str) + 'miniml/'
df.head()
```

```python
#loading the Accession information from efetch table - GSEs

df_gse = pd.read_csv('list_access_efetch.txt', sep=r'\s{2,}', engine='python', header=None, names=['GSE_ID'])


df_gse['GSE'] = df_gse.GSE_ID.str.split('Accession: ', expand=True)[1]
df_gse['GSE'] = df_gse.GSE.str.split("\t", expand=True)[0]

df_gse_index = df_gse.reset_index(drop=True)
df_gse_index.head()
```

```python
df_gse_ftp = pd.concat([df_gse_index, df], axis=1)
df_gse_ftp['GSE'].nunique()
# df_gse_ftp.head()
```

```python
df_gse_ftp_final = df_gse_ftp[['GSE', 'Address']]
df_gse_ftp_final.head()
```

```python

df_gse_ftp_final['Full_Address'] = df_gse_ftp_final['Address'].astype(str) + df_gse_ftp_final['GSE'] + '_family.xml.tgz'

df_gse_ftp_final.head()
```

```python
#saving this dataframe with information about GSE and their respective ftp address
df_gse_ftp_final.to_csv(r'/home/local/USHERBROOKE/frog2901/Documents/XML_TEST/GSE_FTP_Hs_ChipSeq.csv' , index = None, header=True)


```

```python
#subfile from final dataframe

df_gse_ftp_final_test_20rows = df_gse_ftp_final[:20]
df_gse_ftp_final_test_20rows
```

```python
#Get directories

import subprocess
import os

#get the files using the addres column and keep the code to adjust it here


#Funciton create directories to store our files

def create_dir(path):    
    try:
        os.mkdir(path)
        print("Directory {} done".format(path))
    except OSError:
        print("Directory {} failed".format(path))
            

root_path = '/Users/gfrosi/Documents/Frosi_lab_jacques/XML_TEST'
prefix_dir = 'GEO_'
counter = 1                                                                  

path = os.path.join(root_path , prefix_dir + str(counter))
create_dir(path)


#create the correct link and download the xml files

for index, row in df_gse_ftp_final_test_20rows.iterrows():    
    if (index + 1) % 10 == 0:  
        counter += 1
        path = os.path.join(root_path, prefix_dir + str(counter))
        create_dir(path)    
        
    sample_name = row["Address"].split("/")
    link = row["Address"] + sample_name[6] + "_family.xml.tgz" #we can keep in this form, or get the full address
    subprocess.run(["wget", link, "-P", path])
    
```
