```python
import subprocess
import os
```

```python
#teste 2 - optimizing link cration in a loop. It works. We need to create directories for each 500 files downloaded - lets try with a df 

#Testing in df containing 20 rows

#index - number of rows...serves as a count

for index, row in df_ftp_gse_id_test_get_xml_2.iterrows():
    sample_name = row["Address"].split("/")
    test_2 = '/home/local/USHERBROOKE/frog2901/Documents/XML_TEST/test_2' #specifying a path
    link = row["Address"] + sample_name[6] + "_family.xml.tgz"
  
    subprocess.run(["wget", link, "-P", test_2])
    
```

```python
#Get directories

#problems solved after xxxxxxxxx testsss
#the indetation of last code / counter working
#test with %5 in a df with 20 rows


#Funciton create directories to store our files
def create_dir(path):    
    try:
        os.mkdir(path)
        print("Directory {} done".format(path))
    except OSError:
        print("Directory {} failed".format(path))
            

root_path = '/home/local/USHERBROOKE/frog2901/Documents/XML_TEST'
prefix_dir = 'GEO_'
counter = 1                                                                  

path = os.path.join(root_path , prefix_dir + str(counter))
create_dir(path)


#create the correct link and download the xml files
for index, row in df_ftp_gse_xml_20_rows.iterrows():    
    if (index + 1) % 10 == 0:  
        counter += 1
        path = os.path.join(root_path, prefix_dir + str(counter))
        create_dir(path)    
        
    sample_name = row["Address"].split("/")
    link = row["Address"] + sample_name[6] + "_family.xml.tgz" 
    subprocess.run(["wget", link, "-P", path])
    
```
