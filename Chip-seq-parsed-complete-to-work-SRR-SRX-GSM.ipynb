# Get the SRR ID for all GSM from GEO-NCBI


```python
#create a list from GSM colum


gsm_list = df_chipseq_parsed['GSM'].tolist()
print(len(gsm_list))
```

```python
#Get the address to access the pages to webscrapping the SRR numbers related to each GSM from GSE xml files
#gsm plus SRX address

import os.path
from pathlib import Path
import tarfile

base_dir = "/Users/gfrosi/Documents/Frosi_lab_jacques/XML_TEST/GEO-data-xml"

gsm_list_srx_address = []
pathlist = Path(base_dir).glob('**/*.tgz')
count = 0

for path in pathlist:
    t = tarfile.open(path, 'r')
    file_name =t.getnames()

    
    for file in file_name:
        try:
            f = t.extractfile(file)
        except KeyError:
            print('ERROR: Did not find %s in tar archive' % file)
        else:
            for line in f.readlines():
                # convert byte-like object to str
                line = line.decode("utf-8")
                
                if "<Sample iid=" in line:
                    gsm = line.replace('<Sample iid="','')[:-3].strip()
                                       
                    
                    if gsm in gsm_list:
                        count = 1
                        print('OK')
                        
                
                if '<Relation type="SRA" target=' in line and  count == 1:

                    target  = line.split('=')
                    srx = target[2].replace('"', '') + '='+ target[3].replace("/>",'').replace('"', '').strip()
                    gsm_list_srx_address.append([gsm,srx])
                    count = 0
                    
                    
                    
                    
gsm_list_srx_address          

```

```python
#removing duplicates from our pilot list (a GSM could be inserted in more then one GSE)


#We recovery 42220 SRX


no_dup = set(tuple(row) for row in gsm_list_srx_address)
gsm_list_srx_address_nodup = list(no_dup)
print(len(gsm_list_srx_address_nodup))
```

```python
#Creating a file to write our list_srx_srr_test (with GSM related)

gsm_srx_address_file = open("gsm_srx_address_all_table_mac.tsv", "w")
for i in gsm_list_srx_address_nodup: #our list
    line = "\t".join(i)
    line += "\n"
    gsm_srx_address_file.write(line)

gsm_srx_address_file.close()
```

```python
#V1 - SRR and SRX for our control

import urllib.request
import lxml.html as lh
from bs4 import BeautifulSoup
import requests
from time import sleep
from retry import retry


SRR_srx_ncbi =[]

@retry(TimeoutError, tries=5, delay=3)
def srr_web_scrapping(list_srx_address_no_dup):
    for srx in list_srx_address_no_dup:
        url = srx
        response = requests.get(url)
        data = response.text
        soup = BeautifulSoup(data, 'lxml')

        for a in soup.findAll('b'):
            inf = a.get_text()
            srx = inf.split(':')[0]

        for td in soup.findAll("td"):
            for link in td.find_all('a'):

                SRR_srx_ncbi.append([srx, link.get_text()])



    print(SRR_srx_ncbi)
    
srr_web_scrapping(list_srx_address_no_dup)
```

```python
header = ['SRX', 'SRR']

df_1 = pd.DataFrame(SRR_srx_ncbi, columns=header)

df_srx_srr = df_1.groupby('SRX')['SRR'].apply(lambda x: "%s" % ','.join(x)).reset_index()
df_srx_srr
```
