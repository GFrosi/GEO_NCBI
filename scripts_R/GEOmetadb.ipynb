```python
#summary dataframe

def summary(df):
    # Identify the Data Types
    s_df1 = pd.DataFrame(df.dtypes, columns=['Dtype'])
    # Identify the unique values
    s_df1['Nunique'] = df.nunique()
    # Identify the Missing values
    s_df1['MissingValues'] = df.isnull().sum
    # Identify the count for each variable
    s_df1['Count'] = df.count()
    # Identify the zero values for each variable
    s_df1['ZeroValues'] = (df==0).sum()
 
    print(s_df1)
    
    
    
    
summary(df1)
```
