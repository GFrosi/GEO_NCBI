```python
#Loading the 60 file names h3k4me3-h3k4me1

df_srr_name_60 = pd.read_csv(r'/Users/gfrosi/Documents/Frosi_lab_jacques/test_pipeline_encode/BigWig_test/sbatch-h3k4me1.txt',  sep="\t", header=None, names=['srr'])
```

```python
#CREATE a function to get just the srr and add in a new column and return the dataframe 

#split AND getting the srr and storing in a new column 

df_srr_name_60['SRR'] = df_srr_name_60['srr'].str.split("_").str[0]

# Create a dict from a df
dict_srr_name_60 = df_srr_name_60.set_index('srr').to_dict()['SRR']
dict_srr_name_60

```

```python
#code json - optmized - df all informations 
print('{"Datasets": \n\t[')
for index, row in result_60.iterrows():
 
    if index > 0:
        print(",")
        
    if row['srr'] in dict_srr_name_60.keys():
        assay = "h3k4me1"
        fn = row['srr']
        GSE = row['GSE']
        GSM = row['GSM']
        cell_type = ''
        
        if isinstance(row['cell type'], str):
            cell_type = row['cell type']
            
        
        print(f'\t\t{{ "file_name":"{fn}", "assay":"{assay}", "GSE":"{GSE}", "GSM":"{GSM}", "cellType": "{cell_type}", "fraglen": ""}}', end="")

print("\n\t]\n}")
```
