```python
######OPTIMIZATION
def replace_substring_bigwig(list_df, regex:[()], columns:[]):
    
    list_of_df_plot = []
    
    for df in list_df:        
        df_local = df.copy()    
        for col in columns:
            for i in regex:
                 df_local[col]= df_local[col].str.replace(i[0],i[1])
                    
        list_of_df_plot.append(df_local)

    
    return list_of_df_plot

list_of_regex = [(r'.signal-testd.bigwig', ''), (r'.merged.raw.bigwig', ''), (r'\.\w+\.\w{19}\d+', ''), (r'.bigwig', '')]
columns=['File1', 'File2']

list_of_df_plot = replace_substring_bigwig(list_of_raw_df,list_of_regex, columns) 
                                           
                                           
```

```python
list_of_df_plot[7]
```

```python
######OPTIMIZATION
#pivoted and transposed

def pivot_trans(list_of_df_plot):
    
    list_of_piv_t_df = []
    
    for df in list_of_df_plot:
        result = df.pivot_table(index='File1', columns='File2', values='correlation')

        #transposed table
        list_of_piv_t_df.append(result.T)
    
    return list_of_piv_t_df
```

```python
######OPTIMIZATION
list_of_piv_t_df = pivot_trans(list_of_df_plot)
list_of_piv_t_df[7]
```

```python
######OPTIMIZATION
def cluster(df, num):
    f, ax = plt.subplots(figsize=(20, 10))
    sns.heatmap(df, annot=True, fmt=".2g", cmap="BuPu" ,square=True, linewidths=.5, cbar_kws={'label': 'Colorbar'}, annot_kws={'size':16})


    #To avoid the cutted squares at the top and bottom is necessary to add a space

    b, t = plt.ylim() # discover the values for bottom and top
    b += 0.5 # Add 0.5 to the bottom
    t -= 0.5 # Subtract 0.5 from the top
    plt.ylim(b, t) # update the ylim(bottom, top) values


    plt.title("DF_" + str(num), fontsize =20)
    plt.xlabel('Replicates', fontsize = 12)
    plt.ylabel('Replicates', fontsize = 12)
    plt.show()

for i, df_piv_t in enumerate(list_of_piv_t_df):
    cluster(df_piv_t, i + 1)
```
