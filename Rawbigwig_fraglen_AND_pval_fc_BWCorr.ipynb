```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


pd.set_option('display.max_rows', 20)

df_1 = pd.read_csv(r'/Users/gfrosi/Documents/Frosi_lab_jacques/test_pipeline_encode/BigWig_test/out_corr_all_frag_pval_fc_rbw.txt', sep="\t", header=None, names=['File1','File2', 'correlation'])
```

```python
#First - remove orig file 
df_1 = df_1.drop(df_1[df_1['File2'].str.contains("-orig.merged.raw.bigwig")].index)
df_1 = df_1.drop(df_1[df_1['File1'].str.contains("-orig.merged.raw.bigwig")].index)
df_1.reset_index(inplace=True, drop=True)

df_1
```

```python
def replace_substring_bigwig(df, regex:[()], columns:[]):
   
    df_local = df.copy()
#     h3k4me1_df['File1'] = h3k4me1_df['File1'].str.replace(r'^SRR\d+\w+\.\w{15}', 'h3k4me1').str.replace(r'\.bigwig','')
    
    for col in columns:
        for i in regex:
             df_local[col]= df_local[col].str.replace(i[0],i[1]) 
    
    
    return df_local

list_of_regex = [(r'.signal-testd.bigwig', ''), (r'.merged.raw.bigwig', ''), (r'\.\w+\.\w{19}\d+', ''), (r'.bigwig', '')]
columns=['File1', 'File2']


df_1_replaced = replace_substring_bigwig(df_1,list_of_regex, columns) 
                                           
 
```

```python
#Checking 
df_1_replaced
```

```python
#pivoting df


df_piv = df_1_replaced.pivot_table(index='File1', columns='File2', values='correlation')
```

```python
df_piv 
```

```python
#adding the index and column to get a perfect triangular matrix

def add_row_col(df):
    
    df_local = df.copy()
    
        
    index = df_local.index.tolist()
    cols = df_local.columns.tolist()

        #creating a new row 
        
    new_df = df_local.append(pd.Series(name=cols[-1]))
    new_df[index[0]] = np.nan
        
        #creating a new column and change its place 
    new_cols = new_df.columns.tolist()
    new_cols = new_cols[-1:] + new_cols[:-1]

        
    final_df =  new_df[new_cols]
        
        
    return  final_df


df_complete = add_row_col(df_piv)
```

```python
#checking
df_complete
```

```python
#fill_na and mirror

def fill_na_mirror(df):
    
    df_local = df.copy()

    df_zero = df_local.fillna(0)
    df_mirror = df_zero + df_zero.T - np.diag(np.diag(df_zero))

        

    return df_mirror



df_complete_mirror = fill_na_mirror(df_complete)
```

```python
#checking
df_complete_mirror
```

```python
#fill diagonal with 1


def fill_diag(df):
    
    df_local = df.copy()


    list_index = df_local.index.tolist()
    list_columns = df_local.columns.tolist()

    #     print(list_columns)

    for i in list_index:
        for c in list_columns:
            if i == c:
    #                 print(i,c)
                df_local.loc[i,c]=1

        
    return df_local


final_df = fill_diag(df_complete_mirror)
```

```python
#checking
final_df
```

```python
#plotting clustermap 


g = sns.clustermap(final_df, cmap="mako", standard_scale=1, metric="correlation", 
                       cbar_kws={'label': 'Colorbar'}, yticklabels=1, xticklabels=1)
g.ax_heatmap.set_xticklabels(g.ax_heatmap.get_xmajorticklabels(), fontsize = 5)
g.ax_heatmap.set_yticklabels(g.ax_heatmap.get_ymajorticklabels(), fontsize = 5)

fig = plt.gcf()
fig.set_size_inches(15,15)
#sns.set(font_scale=1.4)

plt.gcf().get_axes()[2].set_ylim(final_df.shape[0]-0.01, -0.01)
    #fig.set_ylim(df_mirror.shape[0]-0.5, -0.5)
plt.title("test", fontsize =20)
# sns.set(font_scale=1.25)

# sns.set_context("notebook")

# plt.savefig('plot.png', dpi=300)

plt.show()



```
