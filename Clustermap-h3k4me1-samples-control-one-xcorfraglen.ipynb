```python
#pivoting table
df_piv = df_1.pivot_table(index='File1', columns='File2', values='correlation')
```

```python
df_piv
```

```python
#adding the index and column to get a perfect triangular matrix

def add_row_col(df):
    
    df_local = df.copy()
    
        
    index = df_local.index.tolist()
    cols = df_local.columns.tolist()

        #creating a new row 
        
    new_df = df_local.append(pd.Series(name=cols[-1]))
    new_df[index[0]] = np.nan
        
        #creating a new column and change its place 
    new_cols = new_df.columns.tolist()
    new_cols = new_cols[-1:] + new_cols[:-1]

        
    final_df =  new_df[new_cols]
        
        
    return  final_df


df_complete = add_row_col(df_piv)
```

```python
df_complete
```

```python
#fill_na and mirror

def fill_na_mirror(df):
    
    df_local = df.copy()

    df_zero = df_local.fillna(0)
    df_mirror = df_zero + df_zero.T - np.diag(np.diag(df_zero))

        

    return df_mirror



df_complete_mirror = fill_na_mirror(df_complete)
```

```python
df_complete_mirror
```

```python
#fill diagonal with 1


def fill_diag(df):
    
    df_local = df.copy()


    list_index = df_local.index.tolist()
    list_columns = df_local.columns.tolist()

    #     print(list_columns)

    for i in list_index:
        for c in list_columns:
            if i == c:
    #                 print(i,c)
                df_local.loc[i,c]=1

        
    return df_local



final_df = fill_diag(df_complete_mirror)
```

```python
final_df.iloc[10:]
```

```python
#checking the diagonal...
#the diagonal is in the correct place...

final_df.to_csv(r'h3k4me1-cluster.csv', index=True)
```

```python
#plotting clustermap 

%matplotlib 

g = sns.clustermap(final_df, cmap="mako", metric="correlation", 
                       cbar_kws={'label': 'Colorbar'}, yticklabels=1, xticklabels=1)

# g.ax_heatmap.set_xticklabels(g.ax_heatmap.get_xmajorticklabels(), fontsize = 8)
# g.ax_heatmap.set_yticklabels(g.ax_heatmap.get_ymajorticklabels(), fontsize = 8)


fig = plt.gcf()
fig.set_size_inches(15,15)
#sns.set(font_scale=1.4)

plt.gcf().get_axes()[2].set_ylim(final_df.shape[0]-0.01, -0.01)

plt.title("test h3k4me3", fontsize =20)


plt.show()
```
